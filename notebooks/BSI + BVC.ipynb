```python
import pandas as pd
import numpy as np

from tseries_patterns.buysell import HawkesBSI, HawkesBVC
```

# Data

```python
df = pd.read_csv("csv/volumebars.csv", parse_dates=['stamp'])
```

# BSI

```python
bsi = HawkesBSI(0.1)
bsi.eval (df)
bsi.plot(title="variation with kappa = 0.1")
```

```python
obj = HawkesBSI(0.5)
obj.eval (df)
obj.plot(title="variation with kappa = 0.5")
```

# BVC

```python
obj = HawkesBVC(30,0.1)
obj.eval (df)
obj.plot(title="variation with kappa = 0.1")
```

```python
obj = HawkesBVC(30,0.05)
obj.eval (df)
obj.plot(title="variation with kappa = 0.05")
```
